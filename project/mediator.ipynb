```python
df1 = pd.read_excel('1_插值补充后数据.xlsx')
df = pd.merge(df1, df_final, 
                    left_on=['地级市', '年份_final'], 
                    right_on=['地区', '年份'], 
                    how='left')
# 方法1：直接删
df.drop(['年份_db', '省份_db', '城市', '省份代码_db', '城市代码', '地区', '年份','Unnamed: 0'], axis=1, inplace=True)
#df.to_excel('mediate_merge.xlsx')
```

```python
#构建评分环节
weights = {
    'GDP(亿元)': 0.1601,
    'GDP增长率(%)': 0.0883,
    '人均GDP(元)': 0.0516,
    '供水量(市辖区)(万吨)': 0.0249,
    '医院床位数(张)': 0.0267,
    '单位从业人员数(人)': 0.1165,
    '地方一般公共预算支出中科学技术支出(万元)': 0.1348,
    '外商直接投资额(万美元)': 0.0465,
    '天然气供气总量(市辖区)(万立方米)': 0.0249,
    '工业用电量(亿千瓦时)': 0.0494,
    '普通高等学校在校学生数(人)_y': 0.1005,
    '货物出口额(亿美元)': 0.0381,
    '货物进口额(亿美元)': 0.0381,
    '金融机构各项贷款余额(万元)': 0.0498,
    '金融机构存款余额(亿元)': 0.0498
}
cols = list(weights.keys())
#采用映射到0-1的min_max函数:
def min_max_norm(x):
    if x.max() == x.min():
        return pd.Series(0.5, index=x.index)
    return (x - x.min()) / (x.max() - x.min())
```

```python
df_norm = df.groupby('年份_final')[cols].transform(min_max_norm)
df['营商环境得分'] = df_norm.apply(lambda row: sum(row[col] * weights[col] for col in cols), axis=1)
df['营商环境得分'] = df['营商环境得分'] * 100
```

```python
df = df_clean.copy()
# ==========================================
# 3. 构造文献要求的控制变量
# ==========================================
# 注意：等号右侧的 ['列名'] 需要你对照生成的 "1_插值补充后数据.xlsx" 里的实际基础表头进行替换
df['人力资本水平'] = df['普通高等学校在校学生数(人)_x'] / df['户籍人口(万人)']
df['传统基础设施'] = np.log(df['公路货运量(万吨)'] +1 ) # 加1防止出现log(0)
df['互联网发展水平'] = np.log(df['电信业务收入(万元)'] +1)
df['金融发展水平'] = df['年末金融机构各项贷款余额(万元)'] / df['地区生产总值(万元)']
df['财政压力水平'] = df['地方财政一般预算内支出(万元)'] / df['地方财政一般预算内收入(万元)']
df['科学支出水平'] = np.log(df['科学支出(万元)']+165 )
df['城市发展水平'] = np.log(df['人均地区生产总值(元)']+1)
df['外商投资水平'] = np.log(df['外商投资企业数(个)']+1)
df['交通便捷程度'] = df['高速公路里程(公里)'] / df['户籍人口(万人)']
# 将构造好的控制变量放入列表
controls = ['地区生产总值(万元)','人口密度(人／平方公里)', '第三产业增加值占GDP比重(%)', '人力资本水平', '每百人公共图书馆藏书(册、件)',
            '传统基础设施', '互联网发展水平', '金融发展水平', '财政压力水平', '科学支出水平','城市发展水平','外商投资水平','生活垃圾无害化处理率(%)','交通便捷程度']

# 剔除在构造变量后仍然存在缺失值（如缺乏某城市整体面积数据导致无法插值）的样本
df.dropna(subset=['诉求量', 'post（开放数据平台时间虚拟变量）'] + controls, inplace=True)
#df.to_excel("2_处理后含控制变量数据.xlsx", index=False)
```

```python
# ==========================================
# 4. 双重机器学习 (DML) 基准回归结合中介效应
# ==========================================
df['诉求量_log'] = np.log1p(df['诉求量'])
Y = df['诉求量_log'].values
T = df['post（开放数据平台时间虚拟变量）'].values
X = df[controls].values
M = df['营商环境得分'].values
# 提取城市与年份固定效应，转化为虚拟变量矩阵，作为混杂因素 W
# DML第一阶段会利用机器学习算法剥离W对Y和T的影响，从而获得纯净的因果效应
W_df = pd.get_dummies(df[['地级市', '年份_final']], columns=['地级市', '年份_final'], drop_first=True)
W = W_df.values
```

```python
from econml.dml import LinearDML
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
# --- 准备工作 ---
# 确保 X 中不包含用来计算 M 的原始指标，避免共线性逻辑错误
# 如果 controls 里有 GDP 等，建议在此处剔除
# ==========================================
# 路径 A: T (政策) -> M (营商环境)
# ==========================================
print("正在拟合路径 A (T -> M)...")
est_a = LinearDML(model_y=RandomForestRegressor(n_estimators=100), 
                  model_t=RandomForestClassifier(n_estimators=100), 
                  discrete_treatment=True, cv=5)
est_a.fit(M, T, X=X, W=W)

# 提取系数（等同于方程3中的 gamma_1）
gamma_1 = est_a.intercept_ 
print(f"路径 A (T->M) 的平均效应系数: {gamma_1:.4f}")
print(est_a.summary().tables[0])

# ==========================================
# 路径 B: M (营商环境) -> Y (诉求量)
# ==========================================
print("\n正在拟合路径 B (M -> Y)...")
# 将 T 加入混杂因素 W，以控制政策本身的直接影响
W_with_T = np.hstack([W, T.reshape(-1, 1)])

est_b = LinearDML(model_y=RandomForestRegressor(n_estimators=100), 
                  model_t=RandomForestRegressor(n_estimators=100), 
                  discrete_treatment=False, # M 是连续得分
                  cv=5)
est_b.fit(Y, M, X=X, W=W_with_T)

# 提取系数（等同于方程2中的 beta_2）
beta_2 = est_b.intercept_
print(f"路径 B (M->Y) 的平均效应系数: {beta_2:.4f}")
print(est_b.summary().tables[0])
```

```python
# ==========================================
# 路径 A: T (政策) -> M (营商环境)
# ==========================================
print("正在拟合路径 A (T -> M)...")

est_a = LinearDML(
    model_y=RandomForestRegressor(n_estimators=70, random_state=42,max_depth= 10),
    model_t=RandomForestClassifier(n_estimators=70, random_state=42,max_depth= 10),
    discrete_treatment=True,
    cv=5,
    random_state=42
)

est_a.fit(Y=M, T=T, X=X, W=W)

# 获取政策对营商环境的平均处理效应
ate_a = est_a.intercept_
print(f"路径 A (T→M) 的平均效应: {ate_a:.4f}")

# 查看详细结果
print("\n路径A详细结果:")
print(est_a.summary())

# 异质效应
a_effects = est_a.effect(X)
print(f"效应范围: [{a_effects.min():.4f}, {a_effects.max():.4f}]")

# 提取系数和标准误（用于后续中介效应计算）
a_coef = ate_a
#a_se = est_a.summary().tables[0]['std err'][0]  # 提取标准误
#print(f"\n路径A系数提取: a = {a_coef:.4f}, se(a) = {a_se:.4f}")


# ==========================================
# 路径 B: M (营商环境) -> Y (创业活力)
# ==========================================
print("\n正在拟合路径 B (M -> Y)...")

# 将 T 和原始 W 合并作为控制变量（在控制政策直接效应的条件下估计 M→Y）
if W is not None:
    W_for_b = np.column_stack([W, T])
else:
    W_for_b = T.reshape(-1, 1)

est_b = LinearDML(
    model_y=RandomForestRegressor(n_estimators=70, random_state=42,max_depth= 10),
    model_t=RandomForestRegressor(n_estimators=70, random_state=42,max_depth= 10),  # M 是连续变量
    discrete_treatment=False,  # 关键：M 是连续变量
    cv=5,
    random_state=42
)

est_b.fit(Y=Y, T=M, X=X, W=W_for_b)

# 获取营商环境对创业活力的平均效应
ate_b = est_b.intercept_
print(f"路径 B (M→Y) 的平均效应: {ate_b:.4f}")

# 查看详细结果
print("\n路径B详细结果:")
print(est_b.summary())

# 提取系数和标准误
b_coef = ate_b
b_se = est_b.summary().tables[0]['std err'][0]
print(f"\n路径B系数提取: b = {b_coef:.4f}, se(b) = {b_se:.4f}")
```

```python
est_c = LinearDML(
    model_y=RandomForestRegressor(n_estimators=200, random_state=123,max_depth=15),
    model_t=RandomForestClassifier(n_estimators=200, random_state=123,max_depth=15),
    discrete_treatment=True,
    cv=5,
    random_state=42
)

est_c.fit(Y=Y, T=T, X=X, W=W)

ate_c = est_c.intercept_
print(f"路径 C (T→Y) 的总效应: {ate_c:.4f}")
```

```python
est_cprime = LinearDML(
    model_y=RandomForestRegressor(n_estimators=70, random_state=42,
                                   min_samples_leaf=20, max_depth=5),
    model_t=RandomForestRegressor(n_estimators=70, random_state=42,
                                   min_samples_leaf=20, max_depth=5),  # 多变量处理用回归器
    discrete_treatment=False,  # 连续/多值处理变量设为 False
    cv=5,
    random_state=42
)

# 将 T 和 M 合并成二维数组作为处理变量
treatment = np.column_stack([T, M])

est_cprime.fit(Y=Y, T=treatment, X=X,W=W)
print("\n路径 c' 和 b (T+M → Y):")
print(est_cprime.summary())
c_prime = est_cprime.coef_[0]  # 直接效应 c'
b = est_cprime.coef_[1]       # M → Y 的效应 b
```

```python
# 间接效应（中介效应）= 路径A × 路径B
indirect_effect = ate_a * ate_b
print(f"间接效应 (T→M→Y): {indirect_effect:.4f}")

# 直接效应（T对Y的直接效应）= 总效应 - 间接效应
direct_effect = ate_c - indirect_effect
print(f"直接效应 (T→Y 除去M的中介): {direct_effect:.4f}")
```

```python
# 1. 提取路径 A 的核心系数 (gamma_1: T -> M)
summary_a = est_a.summary().tables[0]
# 在 LinearDML 中，intercept_ 对应的就是处理变量 T 的平均系数
gamma_1_val = est_a.intercept_ 
p_val_a = float(summary_a.data[1][4]) # 提取 P 值

# 2. 提取路径 B 的核心系数 (beta_2: M -> Y)
summary_b = est_b.summary().tables[0]
beta_2_val = est_b.intercept_
p_val_b = float(summary_b.data[1][4])

print(f"--- 中介效应核心指标回顾 ---")
print(f"路径 A (T -> M): 系数 = {gamma_1_val:.4f}, P值 = {p_val_a:.4f}")
print(f"路径 B (M -> Y): 系数 = {beta_2_val:.4f}, P值 = {p_val_b:.4f}")

# 3. 判断中介成立条件
if p_val_a < 0.05 and p_val_b < 0.05:
    mediation_effect = gamma_1_val * beta_2_val
    print(f"结论：中介效应显著！间接效应大小为: {mediation_effect:.4f}")
else:
    print("结论：中介效应不显著，请检查数据质量或理论逻辑。")
```
