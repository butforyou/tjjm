```python
#构建评分环节
weights = {
    'GDP(亿元)': 0.1601,
    'GDP增长率(%)': 0.0883,
    '人均GDP(元)': 0.0516,
    '供水量(市辖区)(万吨)': 0.0249,
    '医院床位数(张)': 0.0267,
    '单位从业人员数(人)': 0.1165,
    '地方一般公共预算支出中科学技术支出(万元)': 0.1348,
    '外商直接投资额(万美元)': 0.0465,
    '天然气供气总量(市辖区)(万立方米)': 0.0249,
    '工业用电量(亿千瓦时)': 0.0494,
    '普通高等学校在校学生数(人)_y': 0.1005,
    '货物出口额(亿美元)': 0.0381,
    '货物进口额(亿美元)': 0.0381,
    '金融机构各项贷款余额(万元)': 0.0498,
    '金融机构存款余额(亿元)': 0.0498
}
cols = list(weights.keys())
#采用映射到0-1的min_max函数:
def min_max_norm(x):
    if x.max() == x.min():
        return pd.Series(0.5, index=x.index)
    return (x - x.min()) / (x.max() - x.min())
```

```python
# ==========================================
# 4. 双重机器学习 (DML) 基准回归结合中介效应
# ==========================================
df['诉求量_log'] = np.log1p(df['诉求量'])
Y = df['诉求量_log'].values
T = df['post（开放数据平台时间虚拟变量）'].values
X = df[controls].values
M = df['营商环境得分'].values
# 提取城市与年份固定效应，转化为虚拟变量矩阵，作为混杂因素 W
# DML第一阶段会利用机器学习算法剥离W对Y和T的影响，从而获得纯净的因果效应
W_df = pd.get_dummies(df[['地级市', '年份_final']], columns=['地级市', '年份_final'], drop_first=True)
W = W_df.values
```

```python
from econml.dml import LinearDML
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
# --- 准备工作 ---
# 确保 X 中不包含用来计算 M 的原始指标，避免共线性逻辑错误
# 如果 controls 里有 GDP 等，建议在此处剔除
# ==========================================
# 路径 A: T (政策) -> M (营商环境)
# ==========================================
print("正在拟合路径 A (T -> M)...")
est_a = LinearDML(model_y=RandomForestRegressor(n_estimators=100), 
                  model_t=RandomForestClassifier(n_estimators=100), 
                  discrete_treatment=True, cv=5)
est_a.fit(M, T, X=X, W=W)

# 提取系数（等同于方程3中的 gamma_1）
gamma_1 = est_a.intercept_ 
print(f"路径 A (T->M) 的平均效应系数: {gamma_1:.4f}")
print(est_a.summary().tables[0])

# ==========================================
# 路径 B: M (营商环境) -> Y (诉求量)
# ==========================================
print("\n正在拟合路径 B (M -> Y)...")
# 将 T 加入混杂因素 W，以控制政策本身的直接影响
W_with_T = np.hstack([W, T.reshape(-1, 1)])

est_b = LinearDML(model_y=RandomForestRegressor(n_estimators=100), 
                  model_t=RandomForestRegressor(n_estimators=100), 
                  discrete_treatment=False, # M 是连续得分
                  cv=5)
est_b.fit(Y, M, X=X, W=W_with_T)

# 提取系数（等同于方程2中的 beta_2）
beta_2 = est_b.intercept_
print(f"路径 B (M->Y) 的平均效应系数: {beta_2:.4f}")
print(est_b.summary().tables[0])
```

```python
import numpy as np
import pandas as pd
from tqdm import tqdm
from econml.dml import LinearDML
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.utils import resample

# ==========================================
# 正确的 Bootstrap 中介效应检验
# ==========================================

def bootstrap_mediation_correct(T, M, Y, X, W, n_bootstrap=100, conf_level=0.95):
    """
    正确的 Bootstrap 中介检验：
    1. 在原样本上估计路径 A 和路径 B 的模型
    2. Bootstrap 重抽样得到系数的分布
    3. 计算间接效应的置信区间
    """
    
    n_samples = len(T)
    np.random.seed(42)
    
    # 第一步：在原样本上拟合模型（获取估计器，但不提取系数）
    print("步骤 1/3: 在原样本上训练基础模型...")
    
    # 路径 A: T -> M
    est_a = LinearDML(
        model_y=RandomForestRegressor(n_estimators=50, random_state=42),
        model_t=RandomForestClassifier(n_estimators=50, random_state=42),
        discrete_treatment=True, cv=3, random_state=42
    )
    est_a.fit(Y=M, T=T, X=X, W=W)
    
    # 路径 B: M -> Y (控制 T)
    if W is not None:
        W_with_T = np.column_stack([W, T])
    else:
        W_with_T = T.reshape(-1, 1)
    
    est_b = LinearDML(
        model_y=RandomForestRegressor(n_estimators=50, random_state=42),
        model_t=RandomForestRegressor(n_estimators=50, random_state=42),
        discrete_treatment=False, cv=3, random_state=42
    )
    est_b.fit(Y=Y, T=M, X=X, W=W_with_T)
    
    # 获取原始点估计
    a_original = est_a.intercept_[0] if hasattr(est_a.intercept_, '__len__') else est_a.intercept_
    b_original = est_b.intercept_[0] if hasattr(est_b.intercept_, '__len__') else est_b.intercept_
    indirect_original = a_original * b_original
    
    print(f"原始估计 - a: {a_original:.6f}, b: {b_original:.6f}, 间接效应: {indirect_original:.6f}")
    
    # 第二步：Bootstrap 重抽样
    print(f"\n步骤 2/3: Bootstrap 重抽样 ({n_bootstrap} 次)...")
    
    a_boot = []
    b_boot = []
    
    for i in tqdm(range(n_bootstrap), desc="Bootstrap 进度"):
        # Bootstrap 抽样（有放回）
        indices = resample(range(n_samples), replace=True, n_samples=n_samples)
        
        T_boot = T[indices]
        M_boot = M[indices]
        Y_boot = Y[indices]
        X_boot = X[indices] if X is not None else None
        W_boot = W[indices] if W is not None else None
        
        try:
            # 在 Bootstrap 样本上重新拟合路径 A
            est_a_boot = LinearDML(
                model_y=RandomForestRegressor(n_estimators=50, random_state=None),  # 每次不同
                model_t=RandomForestClassifier(n_estimators=50, random_state=None),
                discrete_treatment=True, cv=3, random_state=None
            )
            est_a_boot.fit(Y=M_boot, T=T_boot, X=X_boot, W=W_boot)
            a_i = est_a_boot.intercept_[0] if hasattr(est_a_boot.intercept_, '__len__') else est_a_boot.intercept_
            
            # 在 Bootstrap 样本上重新拟合路径 B
            if W_boot is not None:
                W_with_T_boot = np.column_stack([W_boot, T_boot])
            else:
                W_with_T_boot = T_boot.reshape(-1, 1)
            
            est_b_boot = LinearDML(
                model_y=RandomForestRegressor(n_estimators=50, random_state=None),
                model_t=RandomForestRegressor(n_estimators=50, random_state=None),
                discrete_treatment=False, cv=3, random_state=None
            )
            est_b_boot.fit(Y=Y_boot, T=M_boot, X=X_boot, W=W_with_T_boot)
            b_i = est_b_boot.intercept_[0] if hasattr(est_b_boot.intercept_, '__len__') else est_b_boot.intercept_
            
            a_boot.append(a_i)
            b_boot.append(b_i)
            
        except Exception as e:
            continue
    
    # 第三步：计算间接效应的置信区间
    print("\n步骤 3/3: 计算置信区间...")
    
    indirect_boot = np.array(a_boot) * np.array(b_boot)
    a_boot = np.array(a_boot)
    b_boot = np.array(b_boot)
    
    alpha = 1 - conf_level
    lower_pct = 100 * alpha / 2
    upper_pct = 100 * (1 - alpha / 2)
    
    # 计算各种统计量
    def get_stats(boot_samples, original):
        mean = np.mean(boot_samples)
        bias = mean - original
        std = np.std(boot_samples, ddof=1)
        ci_lower = np.percentile(boot_samples, lower_pct)
        ci_upper = np.percentile(boot_samples, upper_pct)
        
        # 百分位置信区间
        p_value = 2 * min(
            np.mean(boot_samples <= 0),
            np.mean(boot_samples >= 0)
        )
        
        return {
            'original': original,
            'mean': mean,
            'bias': bias,
            'std': std,
            'ci_lower': ci_lower,
            'ci_upper': ci_upper,
            'p_value': p_value,
            'significant': ci_lower * ci_upper > 0
        }
    
    # 获取总效应（可选）
    print("计算总效应...")
    est_c = LinearDML(
        model_y=RandomForestRegressor(n_estimators=50, random_state=42),
        model_t=RandomForestClassifier(n_estimators=50, random_state=42),
        discrete_treatment=True, cv=3, random_state=42
    )
    est_c.fit(Y=Y, T=T, X=X, W=W)
    c_original = est_c.intercept_[0] if hasattr(est_c.intercept_, '__len__') else est_c.intercept_
    
    # Bootstrap 总效应（可选，计算量较大）
    c_boot = []
    for i in tqdm(range(min(n_bootstrap, 500)), desc="总效应 Bootstrap"):
        indices = resample(range(n_samples), replace=True, n_samples=n_samples)
        try:
            est_c_boot = LinearDML(
                model_y=RandomForestRegressor(n_estimators=100, random_state=None),
                model_t=RandomForestClassifier(n_estimators=100, random_state=None),
                discrete_treatment=True, cv=5, random_state=None
            )
            est_c_boot.fit(Y=Y[indices], T=T[indices], 
                          X=X[indices] if X is not None else None,
                          W=W[indices] if W is not None else None)
            c_i = est_c_boot.intercept_[0] if hasattr(est_c_boot.intercept_, '__len__') else est_c_boot.intercept_
            c_boot.append(c_i)
        except:
            continue
    c_boot = np.array(c_boot)
    
    # 整理结果
    a_res = get_stats(a_boot, a_original)
    b_res = get_stats(b_boot, b_original)
    indirect_res = get_stats(indirect_boot, indirect_original)
    c_res = get_stats(c_boot, c_original) if len(c_boot) > 0 else None
    
    # 计算中介比例
    if c_res and abs(c_res['original']) > 1e-6:
        ratio_original = indirect_original / c_original
        ratio_boot = indirect_boot / c_boot
        ratio_res = get_stats(ratio_boot, ratio_original)
    else:
        ratio_res = None
    
    # 输出结果
    print("\n" + "="*70)
    print("Bootstrap 中介效应检验结果（修正版）")
    print("="*70)
    print(f"有效 Bootstrap 次数: {len(indirect_boot)}/{n_bootstrap}")
    print(f"置信水平: {conf_level*100:.0f}%\n")
    
    print("【路径系数】")
    print(f"路径 A (T→M) - a 系数: {a_res['original']:.6f}")
    print(f"  95% CI: [{a_res['ci_lower']:.6f}, {a_res['ci_upper']:.6f}]")
    print(f"  偏差: {a_res['bias']:.6f}, 标准误: {a_res['std']:.6f}\n")
    
    print(f"路径 B (M→Y) - b 系数: {b_res['original']:.6f}")
    print(f"  95% CI: [{b_res['ci_lower']:.6f}, {b_res['ci_upper']:.6f}]")
    print(f"  偏差: {b_res['bias']:.6f}, 标准误: {b_res['std']:.6f}\n")
    
    print("【核心：中介效应】")
    print(f"间接效应 (a×b): {indirect_res['original']:.6f}")
    print(f"  95% CI: [{indirect_res['ci_lower']:.6f}, {indirect_res['ci_upper']:.6f}]")
    print(f"  Bootstrap p 值: {indirect_res['p_value']:.4f}")
    print(f"  {'✓ 中介效应显著' if indirect_res['significant'] else '✗ 中介效应不显著'}\n")
    
    if c_res:
        print("【总效应与直接效应】")
        print(f"总效应 (c): {c_res['original']:.6f}")
        print(f"  95% CI: [{c_res['ci_lower']:.6f}, {c_res['ci_upper']:.6f}]")
        print(f"直接效应 (c' = c - a×b): {c_res['original'] - indirect_res['original']:.6f}\n")
        
        if ratio_res:
            print("【中介比例】")
            print(f"比例: {ratio_res['original']*100:.2f}%")
            print(f"  95% CI: [{ratio_res['ci_lower']*100:.2f}%, {ratio_res['ci_upper']*100:.2f}%]\n")
    
    return {
        'a': a_res,
        'b': b_res,
        'indirect': indirect_res,
        'total': c_res,
        'mediation_ratio': ratio_res,
        'boot_samples': {
            'a': a_boot,
            'b': b_boot,
            'indirect': indirect_boot,
            'total': c_boot
        }
    }

# ==========================================
# 执行检验（使用较小的 bootstrap 次数测试）
# ==========================================

# 先用较少的 bootstrap 次数测试（确保能跑通）
print("开始 Bootstrap 中介检验...")
results = bootstrap_mediation_correct(
    T=T, M=M, Y=Y, X=X, W=W,
    n_bootstrap=50,  # 先用 500 次测试，正式用 2000+
    conf_level=0.95
)

# ==========================================
# 可视化
# ==========================================

import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# 间接效应分布
axes[0, 0].hist(results['boot_samples']['indirect'], bins=50, edgecolor='black', alpha=0.7, density=True)
axes[0, 0].axvline(results['indirect']['original'], color='red', linestyle='--', linewidth=2, label=f"原始值: {results['indirect']['original']:.4f}")
axes[0, 0].axvline(results['indirect']['ci_lower'], color='green', linestyle=':', linewidth=2, label=f"95% CI")
axes[0, 0].axvline(results['indirect']['ci_upper'], color='green', linestyle=':', linewidth=2)
axes[0, 0].axvline(0, color='black', linestyle='-', linewidth=1, alpha=0.5)
axes[0, 0].set_title(f'间接效应 (a×b) Bootstrap 分布\n{"显著" if results["indirect"]["significant"] else "不显著"}')
axes[0, 0].set_xlabel('效应值')
axes[0, 0].set_ylabel('密度')
axes[0, 0].legend()

# a 系数分布
axes[0, 1].hist(results['boot_samples']['a'], bins=50, edgecolor='black', alpha=0.7, density=True)
axes[0, 1].axvline(results['a']['original'], color='red', linestyle='--', linewidth=2)
axes[0, 1].axvline(results['a']['ci_lower'], color='green', linestyle=':', linewidth=2)
axes[0, 1].axvline(results['a']['ci_upper'], color='green', linestyle=':', linewidth=2)
axes[0, 1].axvline(0, color='black', linestyle='-', linewidth=1, alpha=0.5)
axes[0, 1].set_title(f'路径 A (T→M) 系数分布\np = {results["a"]["p_value"]:.4f}')
axes[0, 1].set_xlabel('系数值')
axes[0, 1].set_ylabel('密度')

# b 系数分布
axes[1, 0].hist(results['boot_samples']['b'], bins=50, edgecolor='black', alpha=0.7, density=True)
axes[1, 0].axvline(results['b']['original'], color='red', linestyle='--', linewidth=2)
axes[1, 0].axvline(results['b']['ci_lower'], color='green', linestyle=':', linewidth=2)
axes[1, 0].axvline(results['b']['ci_upper'], color='green', linestyle=':', linewidth=2)
axes[1, 0].axvline(0, color='black', linestyle='-', linewidth=1, alpha=0.5)
axes[1, 0].set_title(f'路径 B (M→Y) 系数分布\np = {results["b"]["p_value"]:.4f}')
axes[1, 0].set_xlabel('系数值')
axes[1, 0].set_ylabel('密度')

# 散点图：a vs b
axes[1, 1].scatter(results['boot_samples']['a'], results['boot_samples']['b'], alpha=0.3, s=10)
axes[1, 1].axhline(0, color='black', linestyle='-', linewidth=0.5, alpha=0.5)
axes[1, 1].axvline(0, color='black', linestyle='-', linewidth=0.5, alpha=0.5)
axes[1, 1].scatter(results['a']['original'], results['b']['original'], color='red', s=100, marker='*', label='原始估计')
axes[1, 1].set_xlabel('a 系数 (T→M)')
axes[1, 1].set_ylabel('b 系数 (M→Y)')
axes[1, 1].set_title('Bootstrap 系数联合分布')
axes[1, 1].legend()

plt.tight_layout()
plt.show()
```

```python
est_c = LinearDML(
    model_y=RandomForestRegressor(n_estimators=200, random_state=123,max_depth=15),
    model_t=RandomForestClassifier(n_estimators=200, random_state=123,max_depth=15),
    discrete_treatment=True,
    cv=5,
    random_state=42
)

est_c.fit(Y=Y, T=T, X=X, W=W)

ate_c = est_c.intercept_
print(f"路径 C (T→Y) 的总效应: {ate_c:.4f}")
```

```python
# 间接效应（中介效应）= 路径A × 路径B
indirect_effect = ate_a * ate_b
print(f"间接效应 (T→M→Y): {indirect_effect:.4f}")

# 直接效应（T对Y的直接效应）= 总效应 - 间接效应
direct_effect = ate_c - indirect_effect
print(f"直接效应 (T→Y 除去M的中介): {direct_effect:.4f}")
```

```python
# 1. 提取路径 A 的核心系数 (gamma_1: T -> M)
summary_a = est_a.summary().tables[0]
# 在 LinearDML 中，intercept_ 对应的就是处理变量 T 的平均系数
gamma_1_val = est_a.intercept_ 
p_val_a = float(summary_a.data[1][4]) # 提取 P 值

# 2. 提取路径 B 的核心系数 (beta_2: M -> Y)
summary_b = est_b.summary().tables[0]
beta_2_val = est_b.intercept_
p_val_b = float(summary_b.data[1][4])

print(f"--- 中介效应核心指标回顾 ---")
print(f"路径 A (T -> M): 系数 = {gamma_1_val:.4f}, P值 = {p_val_a:.4f}")
print(f"路径 B (M -> Y): 系数 = {beta_2_val:.4f}, P值 = {p_val_b:.4f}")

# 3. 判断中介成立条件
if p_val_a < 0.05 and p_val_b < 0.05:
    mediation_effect = gamma_1_val * beta_2_val
    print(f"结论：中介效应显著！间接效应大小为: {mediation_effect:.4f}")
else:
    print("结论：中介效应不显著，请检查数据质量或理论逻辑。")
```
